```python
# 데이터프레임을 딕셔너리 형태의 리스트로 변경
corpus = []
for index, row in news_data.iterrows():
    news_dict = {
        "title": row['preprocessed_title'],
        "link": row['관련 href'],
        "content": row['preprocessed_content'],
        'content original': row['뉴스 본문'],
        "뉴스 출처": row['뉴스 출처'],
        "날짜": row['날짜'],
        "구분": row['구분'],
        "시간": row['시간'],
        "년월": row['년월'],
    }
    corpus.append(news_dict)

# 결과 확인
print(corpus[0])
```

```python
embedder = SentenceTransformer("jhgan/ko-sbert-sts")
```

```python
embedder
```

```python
# 단어 임베딩 😎
titles_list = [news["title"] for news in corpus]
content_list = [news["content"] for news in corpus]

# Generate embeddings for titles and content separately
titles_list_embeddings = embedder.encode(titles_list, convert_to_tensor=True)
content_embeddings = embedder.encode(content_list, convert_to_tensor=True)
```

```python
corpus_embeddings = torch.cat((titles_list_embeddings, content_embeddings), dim=1)
corpus_embeddings.size()
```

```python
import re

def preprocess_news_title(news_title):

    ''' 😀 News Title pre-processing '''

    # 특수문자 및 [ ] 안의 글자 제거를 위한 정규표현식 패턴
    pattern = r"\[.*?\]"
    cleaned_title = re.sub(pattern, " ", news_title)

    # 특수문자 및 ( ) 안의 글자 제거를 위한 정규표현식 패턴
    pattern = r"\([^)]*[ㄱ-ㅎㅏ-ㅣ가-힣]+[^)]*\)"
    cleaned_title = re.sub(pattern, " ", cleaned_title)


    # '…'를 ' ' (공백)으로 변경
    cleaned_title = cleaned_title.replace('…', ' ')

    # 나머지 특수기호 제거 (숫자와 .은 제외)
    pattern = r"(?<=\d)[^\w\s,%↓↑0-9.()](?=\d)"
    cleaned_title = re.sub(pattern, " ", cleaned_title)

    # 뉴스 제목 끝에 'NH', 'SK', '신한', '하나'와 같은 글자들 제거
    # cleaned_title = re.sub(r"(NH|SK|신한|하나)\s*$", "", cleaned_title)

    # 앞과 뒤의 공백 제거
    cleaned_title = cleaned_title.strip()

    # '  '를 ' ' (공백)으로 변경
    cleaned_title = cleaned_title.replace('...', ' ')
    cleaned_title = cleaned_title.replace('…', ' ')
    cleaned_title = cleaned_title.replace('"', ' ')
    cleaned_title = cleaned_title.replace("'", ' ')
    cleaned_title = cleaned_title.replace('[..', ' ')
    cleaned_title = cleaned_title.replace('“', ' ')
    cleaned_title = cleaned_title.replace('”', ' ')
    cleaned_title = cleaned_title.replace('‘', ' ')
    cleaned_title = cleaned_title.replace('’', ' ')
    cleaned_title = cleaned_title.replace('  ', ' ')
    cleaned_title = cleaned_title.replace('  ', ' ')


    print("뉴스 제목 원본:", news_title)
    print("😀 전처리 후 뉴스 제목:", cleaned_title)

    return cleaned_title

# 뉴스 제목
news_title = "코스닥 10.06P 오른 939.96 마감(1.08%↑)"

# 뉴스 제목 전처리
cleaned_title = preprocess_news_title(news_title)
```

```python
import re

def preprocess_news_content(news_content):

    # (서울=연합뉴스)와 같은 형태의 문자열 제거
    # pattern = r".\([^)]*서울=연합뉴스[^)]*\)"
    # cleaned_content = re.sub(pattern, "", news_content)

    # 이메일 주소 뒤에 나오는 문장을 삭제합니다.
    pattern = r"\b\w+@\w+\.\w+\b.*$"
    cleaned_content = re.sub(pattern, "", news_content)

    # 이메일 주소 제거
    pattern = r"\b\S+@\S+\b"
    cleaned_content = re.sub(pattern, "", cleaned_content)

    # 특수문자 및 [ ] 안의 글자 제거를 위한 정규표현식 패턴
    pattern = r"\[.*?\]"
    cleaned_content = re.sub(pattern, "", cleaned_content)

    # 특수문자 및 ( ) 안의 글자 제거를 위한 정규표현식 패턴
    pattern = r"\([^)]*[ㄱ-ㅎㅏ-ㅣ가-힣]+[^)]*\)"
    cleaned_content = re.sub(pattern, " ", cleaned_content)

    # 나머지 특수기호 제거 (숫자와 .은 제외)
    pattern = r"(?<=\d)[^\w\s㎡㎞↓↑→0-9.0-9,](?=\d)"
    cleaned_content = re.sub(pattern, " ", cleaned_content)

    # 기자 이름 제거
    pattern = r"\b[가-힣]+\s[가-힣]+\s기자 =\s"
    cleaned_content = re.sub(pattern, "", cleaned_content)

    pattern = r"\b[가-힣]+\s기자 =\s"
    cleaned_content = re.sub(pattern, "", cleaned_content)

    # 뉴스 이름 제거
    pattern = r"\b[가-힣]+\뉴스\b\s*"
    cleaned_content = re.sub(pattern, "", cleaned_content)

    # 특수기호 추가 제거
    pattern = r"【[^】]*】"
    cleaned_content = re.sub(pattern, "", cleaned_content)

    # "아시아경제"나 "서울경제" 등 뒤에 나오는 문장을 삭제
    pattern = r"\b(아시아경제|서울경제|한경닷컴|3분증시연합뉴스TV|매일경제|머니투데이|한국경제TV|이데일리|한국경제|연합뉴스)\b.*$"
    cleaned_content = re.sub(pattern, "", cleaned_content)

    # Remove text that comes after "서울경제"
    pattern = r"서울경제.*"
    cleaned_content = re.sub(pattern, "", cleaned_content)

    # #NH투자증권 뒤에 나오는 문장을 삭제
    pattern = r"#NH투자증권\s.*$"
    cleaned_content = re.sub(pattern, "", cleaned_content)

    # 기자 이름과 언론사를 함께 지우기
    pattern = r"\b[가-힣]+\s[가-힣]+\s특파원\s*=\s*"
    cleaned_content = re.sub(pattern, "", cleaned_content)

    # 기자 이름과 언론사를 함께 지우기
    pattern = r"\s[가-힣]+\s특파원\s*=\s*"
    cleaned_content = re.sub(pattern, "", cleaned_content)

    # "관련뉴스해당" 이후의 문장을 삭제
    pattern = r"\b관련뉴스해당\b.*$"
    cleaned_content = re.sub(pattern, "", cleaned_content)

    # '  '를 ' ' (공백)으로 변경
    cleaned_content = cleaned_content.replace('...', ' ')
    cleaned_content = cleaned_content.replace('…', ' ')
    cleaned_content = cleaned_content.replace('■', ' ')
    cleaned_content = cleaned_content.replace('△', ' ')
    cleaned_content = cleaned_content.replace('▲', ' ')
    cleaned_content = cleaned_content.replace('◆', ' ')
    cleaned_content = cleaned_content.replace('"', ' ')
    cleaned_content = cleaned_content.replace("'", ' ')
    cleaned_content = cleaned_content.replace('“', ' ')
    cleaned_content = cleaned_content.replace('”', ' ')
    cleaned_content = cleaned_content.replace('‘', ' ')
    cleaned_content = cleaned_content.replace('’', ' ')
    cleaned_content = cleaned_content.replace('(서울=연합뉴스)', ' ')
    cleaned_content = cleaned_content.replace('·', ' ')
    cleaned_content = cleaned_content.replace('  ', ' ')
    cleaned_content = cleaned_content.replace('  ', ' ')


    # 마침표(".")가 뒤에 오는 경우에는 제거하지 않음
    if cleaned_content.endswith("."):
        cleaned_content = cleaned_content[:-1]

    # 앞과 뒤의 공백 제거
    cleaned_content = cleaned_content.strip()

    # Add a period at the end if not present
    if not cleaned_content.endswith("."):
        cleaned_content += "."

    return cleaned_content

# 뉴스 본문
news_content = """
은행 판매 홍콩H지수 ELS 40억 손실5대은행 내년 만기물량 13조 넘어주요 은행에서 판매한 홍콩H지수 기초 주가연계증권(ELS) 상품에서 약 40억원대 원금손실이 발생한 것으로 드러났다. 문제는 내년까지 만기도래하는 관련 상품 규모가 5대 은행 에서만 13조원 이상으로, 손실 규모는 눈덩이처럼 커질 수 있다는 것이다.31일 금융권에 따르면 국내 시중은행 한 곳에서 판매된 홍콩H지수 기초 ELS 상품에서 이달 원금 손실이 발생했다.ELS는 기초자산으로 삼은 지수 및 개별 종목의 주가와 연계돼 수익구조가 결정되는 파생상품이다. 보통 출시 이후 3년이 지나면 만기일이 도래하며 6개월마다 기초자산 가격을 평가해 조기상환 기회도 준다.이달 만기 도래 규모는 약 103억원이며 손실 예상금액은 약 40억3000만원으로, 손실률이 40% 수준이다. 이번에 원금 손실이 발생한 이유는 상품이 판매된 2021년 이후 홍콩H지수가 약세를 지속했기 때문이다.홍콩H지수는 홍콩 증시에 상장된 중국 기업 가운데 50개 종목을 추려서 산출하는 지수로, 변동성이 높은 것이 특징이다.홍콩H지수는 지난 2021년 2월 1만2000선을 돌파했으나 그해 말 8000대까지 떨어진 뒤 현재 6000대에서 횡보하고 있다. 지난해 10월 말에는 5000대가 붕괴되기도 했다.문제는 내년이다.내년 중 만기가 돌아오는 홍콩H지수 ELS 물량은 13조 5000억원이 넘는다. 주가가 고점이던 지난 2021년 ELS가 대거 발행된 영향이다.금융권 관계자는 ELS는 상대적으로 고연령대가 투자를 많이 했다 면서 올해 하반기에 도래하는 홍콩 H지수 추종 ELS들은 주가 하락으로 조기상환 되지 못한 채 만기를 맞을 가능성이 높다 면서 만기 도래 시점까지 H지수가 고전을 면치 못한다면 손실상환 위험에 그대로 노출될 수 밖에 없다 고 우려했다.그는 이어 요즘 은행권에서는 홍콩 H지수를 포함한 상품의 경우 리스크 때문에 판매를 줄이는 경향이 뚜렷해 최근 발행은 많지 않은 실정 이라고 덧붙였다.한국예탁결제원에 따르면 지난 6월 기준 홍콩H지수를 기초자산으로 하는 ELS 미상환 잔액은 20조6867억원에 달한다.통상 손실 발생 구간 기준선은 최초 기준가격의 50 55% 선에서, 조기상환 기준선은 60 70% 선에서 형성된다는 점을 감안하면 상당수 상품이 녹인 배리어에 진입, 현재도 조기상환 평가가격을 밑돌고 있을 가능성이 있다.
"""
cleaned_content = preprocess_news_content(news_content)
```

```python
from summarizer import Summarizer

def recommend_news(title_query, content_query, top_k=4):
    title_query_embedding = embedder.encode(title_query, convert_to_tensor=True)
    content_query_embedding = embedder.encode(content_query, convert_to_tensor=True)

    query_embedding = torch.cat((title_query_embedding, content_query_embedding), dim=0)

    cos_scores = util.pytorch_cos_sim(query_embedding, corpus_embeddings)[0]
    cos_scores = cos_scores.cpu()

    # 유사도가 100%인 경우(완전히 동일한 뉴스)를 찾아 인덱스를 저장
    same_news_indices = [idx for idx, score in enumerate(cos_scores) if abs(score - 1.0) < 0.00001]

    recommended_news = []
    for idx in range(len(cos_scores)):
        if idx not in same_news_indices:

            recommended_news.append({"title": corpus[idx]["title"].strip(),
                                     "link": corpus[idx]["link"],
                                     "score": cos_scores[idx].item(),
                                     "content": corpus[idx]["content"].strip(),
                                     'content original' : corpus[idx]['content original']
                                     })

    # 유사도(score)를 기준으로 추천 뉴스를 내림차순으로 정렬
    recommended_news = sorted(recommended_news, key=lambda x: x["score"], reverse=True)

    return recommended_news[:top_k]
```
