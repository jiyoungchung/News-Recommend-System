```python
import re

def preprocess_news_title(news_title):

    ''' ğŸ˜€ News Title pre-processing '''

    # íŠ¹ìˆ˜ë¬¸ì ë° [ ] ì•ˆì˜ ê¸€ì ì œê±°ë¥¼ ìœ„í•œ ì •ê·œí‘œí˜„ì‹ íŒ¨í„´
    pattern = r"\[.*?\]"
    cleaned_title = re.sub(pattern, " ", news_title)

    # íŠ¹ìˆ˜ë¬¸ì ë° ( ) ì•ˆì˜ ê¸€ì ì œê±°ë¥¼ ìœ„í•œ ì •ê·œí‘œí˜„ì‹ íŒ¨í„´
    pattern = r"\([^)]*[ã„±-ã…ã…-ã…£ê°€-í£]+[^)]*\)"
    cleaned_title = re.sub(pattern, " ", cleaned_title)


    # 'â€¦'ë¥¼ ' ' (ê³µë°±)ìœ¼ë¡œ ë³€ê²½
    cleaned_title = cleaned_title.replace('â€¦', ' ')

    # ë‚˜ë¨¸ì§€ íŠ¹ìˆ˜ê¸°í˜¸ ì œê±° (ìˆ«ìì™€ .ì€ ì œì™¸)
    pattern = r"(?<=\d)[^\w\s,%â†“â†‘0-9.()](?=\d)"
    cleaned_title = re.sub(pattern, " ", cleaned_title)

    # ë‰´ìŠ¤ ì œëª© ëì— 'NH', 'SK', 'ì‹ í•œ', 'í•˜ë‚˜'ì™€ ê°™ì€ ê¸€ìë“¤ ì œê±°
    # cleaned_title = re.sub(r"(NH|SK|ì‹ í•œ|í•˜ë‚˜)\s*$", "", cleaned_title)

    # ì•ê³¼ ë’¤ì˜ ê³µë°± ì œê±°
    cleaned_title = cleaned_title.strip()

    # '  'ë¥¼ ' ' (ê³µë°±)ìœ¼ë¡œ ë³€ê²½
    cleaned_title = cleaned_title.replace('...', ' ')
    cleaned_title = cleaned_title.replace('â€¦', ' ')
    cleaned_title = cleaned_title.replace('"', ' ')
    cleaned_title = cleaned_title.replace("'", ' ')
    cleaned_title = cleaned_title.replace('[..', ' ')
    cleaned_title = cleaned_title.replace('â€œ', ' ')
    cleaned_title = cleaned_title.replace('â€', ' ')
    cleaned_title = cleaned_title.replace('â€˜', ' ')
    cleaned_title = cleaned_title.replace('â€™', ' ')
    cleaned_title = cleaned_title.replace('  ', ' ')
    cleaned_title = cleaned_title.replace('  ', ' ')


    print("ë‰´ìŠ¤ ì œëª© ì›ë³¸:", news_title)
    print("ğŸ˜€ ì „ì²˜ë¦¬ í›„ ë‰´ìŠ¤ ì œëª©:", cleaned_title)

    return cleaned_title

# ë‰´ìŠ¤ ì œëª©
news_title = "ì½”ìŠ¤ë‹¥ 10.06P ì˜¤ë¥¸ 939.96 ë§ˆê°(1.08%â†‘)"

# ë‰´ìŠ¤ ì œëª© ì „ì²˜ë¦¬
cleaned_title = preprocess_news_title(news_title)
```

```python
import re

def preprocess_news_content(news_content):

    # (ì„œìš¸=ì—°í•©ë‰´ìŠ¤)ì™€ ê°™ì€ í˜•íƒœì˜ ë¬¸ìì—´ ì œê±°
    # pattern = r".\([^)]*ì„œìš¸=ì—°í•©ë‰´ìŠ¤[^)]*\)"
    # cleaned_content = re.sub(pattern, "", news_content)

    # ì´ë©”ì¼ ì£¼ì†Œ ë’¤ì— ë‚˜ì˜¤ëŠ” ë¬¸ì¥ì„ ì‚­ì œí•©ë‹ˆë‹¤.
    pattern = r"\b\w+@\w+\.\w+\b.*$"
    cleaned_content = re.sub(pattern, "", news_content)

    # ì´ë©”ì¼ ì£¼ì†Œ ì œê±°
    pattern = r"\b\S+@\S+\b"
    cleaned_content = re.sub(pattern, "", cleaned_content)

    # íŠ¹ìˆ˜ë¬¸ì ë° [ ] ì•ˆì˜ ê¸€ì ì œê±°ë¥¼ ìœ„í•œ ì •ê·œí‘œí˜„ì‹ íŒ¨í„´
    pattern = r"\[.*?\]"
    cleaned_content = re.sub(pattern, "", cleaned_content)

    # íŠ¹ìˆ˜ë¬¸ì ë° ( ) ì•ˆì˜ ê¸€ì ì œê±°ë¥¼ ìœ„í•œ ì •ê·œí‘œí˜„ì‹ íŒ¨í„´
    pattern = r"\([^)]*[ã„±-ã…ã…-ã…£ê°€-í£]+[^)]*\)"
    cleaned_content = re.sub(pattern, " ", cleaned_content)

    # ë‚˜ë¨¸ì§€ íŠ¹ìˆ˜ê¸°í˜¸ ì œê±° (ìˆ«ìì™€ .ì€ ì œì™¸)
    pattern = r"(?<=\d)[^\w\sã¡ãâ†“â†‘â†’0-9.0-9,](?=\d)"
    cleaned_content = re.sub(pattern, " ", cleaned_content)

    # ê¸°ì ì´ë¦„ ì œê±°
    pattern = r"\b[ê°€-í£]+\s[ê°€-í£]+\sê¸°ì =\s"
    cleaned_content = re.sub(pattern, "", cleaned_content)

    pattern = r"\b[ê°€-í£]+\sê¸°ì =\s"
    cleaned_content = re.sub(pattern, "", cleaned_content)

    # ë‰´ìŠ¤ ì´ë¦„ ì œê±°
    pattern = r"\b[ê°€-í£]+\ë‰´ìŠ¤\b\s*"
    cleaned_content = re.sub(pattern, "", cleaned_content)

    # íŠ¹ìˆ˜ê¸°í˜¸ ì¶”ê°€ ì œê±°
    pattern = r"ã€[^ã€‘]*ã€‘"
    cleaned_content = re.sub(pattern, "", cleaned_content)

    # "ì•„ì‹œì•„ê²½ì œ"ë‚˜ "ì„œìš¸ê²½ì œ" ë“± ë’¤ì— ë‚˜ì˜¤ëŠ” ë¬¸ì¥ì„ ì‚­ì œ
    pattern = r"\b(ì•„ì‹œì•„ê²½ì œ|ì„œìš¸ê²½ì œ|í•œê²½ë‹·ì»´|3ë¶„ì¦ì‹œì—°í•©ë‰´ìŠ¤TV|ë§¤ì¼ê²½ì œ|ë¨¸ë‹ˆíˆ¬ë°ì´|í•œêµ­ê²½ì œTV|ì´ë°ì¼ë¦¬|í•œêµ­ê²½ì œ|ì—°í•©ë‰´ìŠ¤)\b.*$"
    cleaned_content = re.sub(pattern, "", cleaned_content)

    # Remove text that comes after "ì„œìš¸ê²½ì œ"
    pattern = r"ì„œìš¸ê²½ì œ.*"
    cleaned_content = re.sub(pattern, "", cleaned_content)

    # #NHíˆ¬ìì¦ê¶Œ ë’¤ì— ë‚˜ì˜¤ëŠ” ë¬¸ì¥ì„ ì‚­ì œ
    pattern = r"#NHíˆ¬ìì¦ê¶Œ\s.*$"
    cleaned_content = re.sub(pattern, "", cleaned_content)

    # ê¸°ì ì´ë¦„ê³¼ ì–¸ë¡ ì‚¬ë¥¼ í•¨ê»˜ ì§€ìš°ê¸°
    pattern = r"\b[ê°€-í£]+\s[ê°€-í£]+\síŠ¹íŒŒì›\s*=\s*"
    cleaned_content = re.sub(pattern, "", cleaned_content)

    # ê¸°ì ì´ë¦„ê³¼ ì–¸ë¡ ì‚¬ë¥¼ í•¨ê»˜ ì§€ìš°ê¸°
    pattern = r"\s[ê°€-í£]+\síŠ¹íŒŒì›\s*=\s*"
    cleaned_content = re.sub(pattern, "", cleaned_content)

    # "ê´€ë ¨ë‰´ìŠ¤í•´ë‹¹" ì´í›„ì˜ ë¬¸ì¥ì„ ì‚­ì œ
    pattern = r"\bê´€ë ¨ë‰´ìŠ¤í•´ë‹¹\b.*$"
    cleaned_content = re.sub(pattern, "", cleaned_content)

    # '  'ë¥¼ ' ' (ê³µë°±)ìœ¼ë¡œ ë³€ê²½
    cleaned_content = cleaned_content.replace('...', ' ')
    cleaned_content = cleaned_content.replace('â€¦', ' ')
    cleaned_content = cleaned_content.replace('â– ', ' ')
    cleaned_content = cleaned_content.replace('â–³', ' ')
    cleaned_content = cleaned_content.replace('â–²', ' ')
    cleaned_content = cleaned_content.replace('â—†', ' ')
    cleaned_content = cleaned_content.replace('"', ' ')
    cleaned_content = cleaned_content.replace("'", ' ')
    cleaned_content = cleaned_content.replace('â€œ', ' ')
    cleaned_content = cleaned_content.replace('â€', ' ')
    cleaned_content = cleaned_content.replace('â€˜', ' ')
    cleaned_content = cleaned_content.replace('â€™', ' ')
    cleaned_content = cleaned_content.replace('(ì„œìš¸=ì—°í•©ë‰´ìŠ¤)', ' ')
    cleaned_content = cleaned_content.replace('Â·', ' ')
    cleaned_content = cleaned_content.replace('  ', ' ')
    cleaned_content = cleaned_content.replace('  ', ' ')


    # ë§ˆì¹¨í‘œ(".")ê°€ ë’¤ì— ì˜¤ëŠ” ê²½ìš°ì—ëŠ” ì œê±°í•˜ì§€ ì•ŠìŒ
    if cleaned_content.endswith("."):
        cleaned_content = cleaned_content[:-1]

    # ì•ê³¼ ë’¤ì˜ ê³µë°± ì œê±°
    cleaned_content = cleaned_content.strip()

    # Add a period at the end if not present
    if not cleaned_content.endswith("."):
        cleaned_content += "."

    return cleaned_content

# ë‰´ìŠ¤ ë³¸ë¬¸
news_content = """
ì€í–‰ íŒë§¤ í™ì½©Hì§€ìˆ˜ ELS 40ì–µ ì†ì‹¤5ëŒ€ì€í–‰ ë‚´ë…„ ë§Œê¸°ë¬¼ëŸ‰ 13ì¡° ë„˜ì–´ì£¼ìš” ì€í–‰ì—ì„œ íŒë§¤í•œ í™ì½©Hì§€ìˆ˜ ê¸°ì´ˆ ì£¼ê°€ì—°ê³„ì¦ê¶Œ(ELS) ìƒí’ˆì—ì„œ ì•½ 40ì–µì›ëŒ€ ì›ê¸ˆì†ì‹¤ì´ ë°œìƒí•œ ê²ƒìœ¼ë¡œ ë“œëŸ¬ë‚¬ë‹¤. ë¬¸ì œëŠ” ë‚´ë…„ê¹Œì§€ ë§Œê¸°ë„ë˜í•˜ëŠ” ê´€ë ¨ ìƒí’ˆ ê·œëª¨ê°€ 5ëŒ€ ì€í–‰ ì—ì„œë§Œ 13ì¡°ì› ì´ìƒìœ¼ë¡œ, ì†ì‹¤ ê·œëª¨ëŠ” ëˆˆë©ì´ì²˜ëŸ¼ ì»¤ì§ˆ ìˆ˜ ìˆë‹¤ëŠ” ê²ƒì´ë‹¤.31ì¼ ê¸ˆìœµê¶Œì— ë”°ë¥´ë©´ êµ­ë‚´ ì‹œì¤‘ì€í–‰ í•œ ê³³ì—ì„œ íŒë§¤ëœ í™ì½©Hì§€ìˆ˜ ê¸°ì´ˆ ELS ìƒí’ˆì—ì„œ ì´ë‹¬ ì›ê¸ˆ ì†ì‹¤ì´ ë°œìƒí–ˆë‹¤.ELSëŠ” ê¸°ì´ˆìì‚°ìœ¼ë¡œ ì‚¼ì€ ì§€ìˆ˜ ë° ê°œë³„ ì¢…ëª©ì˜ ì£¼ê°€ì™€ ì—°ê³„ë¼ ìˆ˜ìµêµ¬ì¡°ê°€ ê²°ì •ë˜ëŠ” íŒŒìƒìƒí’ˆì´ë‹¤. ë³´í†µ ì¶œì‹œ ì´í›„ 3ë…„ì´ ì§€ë‚˜ë©´ ë§Œê¸°ì¼ì´ ë„ë˜í•˜ë©° 6ê°œì›”ë§ˆë‹¤ ê¸°ì´ˆìì‚° ê°€ê²©ì„ í‰ê°€í•´ ì¡°ê¸°ìƒí™˜ ê¸°íšŒë„ ì¤€ë‹¤.ì´ë‹¬ ë§Œê¸° ë„ë˜ ê·œëª¨ëŠ” ì•½ 103ì–µì›ì´ë©° ì†ì‹¤ ì˜ˆìƒê¸ˆì•¡ì€ ì•½ 40ì–µ3000ë§Œì›ìœ¼ë¡œ, ì†ì‹¤ë¥ ì´ 40% ìˆ˜ì¤€ì´ë‹¤. ì´ë²ˆì— ì›ê¸ˆ ì†ì‹¤ì´ ë°œìƒí•œ ì´ìœ ëŠ” ìƒí’ˆì´ íŒë§¤ëœ 2021ë…„ ì´í›„ í™ì½©Hì§€ìˆ˜ê°€ ì•½ì„¸ë¥¼ ì§€ì†í–ˆê¸° ë•Œë¬¸ì´ë‹¤.í™ì½©Hì§€ìˆ˜ëŠ” í™ì½© ì¦ì‹œì— ìƒì¥ëœ ì¤‘êµ­ ê¸°ì—… ê°€ìš´ë° 50ê°œ ì¢…ëª©ì„ ì¶”ë ¤ì„œ ì‚°ì¶œí•˜ëŠ” ì§€ìˆ˜ë¡œ, ë³€ë™ì„±ì´ ë†’ì€ ê²ƒì´ íŠ¹ì§•ì´ë‹¤.í™ì½©Hì§€ìˆ˜ëŠ” ì§€ë‚œ 2021ë…„ 2ì›” 1ë§Œ2000ì„ ì„ ëŒíŒŒí–ˆìœ¼ë‚˜ ê·¸í•´ ë§ 8000ëŒ€ê¹Œì§€ ë–¨ì–´ì§„ ë’¤ í˜„ì¬ 6000ëŒ€ì—ì„œ íš¡ë³´í•˜ê³  ìˆë‹¤. ì§€ë‚œí•´ 10ì›” ë§ì—ëŠ” 5000ëŒ€ê°€ ë¶•ê´´ë˜ê¸°ë„ í–ˆë‹¤.ë¬¸ì œëŠ” ë‚´ë…„ì´ë‹¤.ë‚´ë…„ ì¤‘ ë§Œê¸°ê°€ ëŒì•„ì˜¤ëŠ” í™ì½©Hì§€ìˆ˜ ELS ë¬¼ëŸ‰ì€ 13ì¡° 5000ì–µì›ì´ ë„˜ëŠ”ë‹¤. ì£¼ê°€ê°€ ê³ ì ì´ë˜ ì§€ë‚œ 2021ë…„ ELSê°€ ëŒ€ê±° ë°œí–‰ëœ ì˜í–¥ì´ë‹¤.ê¸ˆìœµê¶Œ ê´€ê³„ìëŠ” ELSëŠ” ìƒëŒ€ì ìœ¼ë¡œ ê³ ì—°ë ¹ëŒ€ê°€ íˆ¬ìë¥¼ ë§ì´ í–ˆë‹¤ ë©´ì„œ ì˜¬í•´ í•˜ë°˜ê¸°ì— ë„ë˜í•˜ëŠ” í™ì½© Hì§€ìˆ˜ ì¶”ì¢… ELSë“¤ì€ ì£¼ê°€ í•˜ë½ìœ¼ë¡œ ì¡°ê¸°ìƒí™˜ ë˜ì§€ ëª»í•œ ì±„ ë§Œê¸°ë¥¼ ë§ì„ ê°€ëŠ¥ì„±ì´ ë†’ë‹¤ ë©´ì„œ ë§Œê¸° ë„ë˜ ì‹œì ê¹Œì§€ Hì§€ìˆ˜ê°€ ê³ ì „ì„ ë©´ì¹˜ ëª»í•œë‹¤ë©´ ì†ì‹¤ìƒí™˜ ìœ„í—˜ì— ê·¸ëŒ€ë¡œ ë…¸ì¶œë  ìˆ˜ ë°–ì— ì—†ë‹¤ ê³  ìš°ë ¤í–ˆë‹¤.ê·¸ëŠ” ì´ì–´ ìš”ì¦˜ ì€í–‰ê¶Œì—ì„œëŠ” í™ì½© Hì§€ìˆ˜ë¥¼ í¬í•¨í•œ ìƒí’ˆì˜ ê²½ìš° ë¦¬ìŠ¤í¬ ë•Œë¬¸ì— íŒë§¤ë¥¼ ì¤„ì´ëŠ” ê²½í–¥ì´ ëšœë ·í•´ ìµœê·¼ ë°œí–‰ì€ ë§ì§€ ì•Šì€ ì‹¤ì • ì´ë¼ê³  ë§ë¶™ì˜€ë‹¤.í•œêµ­ì˜ˆíƒê²°ì œì›ì— ë”°ë¥´ë©´ ì§€ë‚œ 6ì›” ê¸°ì¤€ í™ì½©Hì§€ìˆ˜ë¥¼ ê¸°ì´ˆìì‚°ìœ¼ë¡œ í•˜ëŠ” ELS ë¯¸ìƒí™˜ ì”ì•¡ì€ 20ì¡°6867ì–µì›ì— ë‹¬í•œë‹¤.í†µìƒ ì†ì‹¤ ë°œìƒ êµ¬ê°„ ê¸°ì¤€ì„ ì€ ìµœì´ˆ ê¸°ì¤€ê°€ê²©ì˜ 50 55% ì„ ì—ì„œ, ì¡°ê¸°ìƒí™˜ ê¸°ì¤€ì„ ì€ 60 70% ì„ ì—ì„œ í˜•ì„±ëœë‹¤ëŠ” ì ì„ ê°ì•ˆí•˜ë©´ ìƒë‹¹ìˆ˜ ìƒí’ˆì´ ë…¹ì¸ ë°°ë¦¬ì–´ì— ì§„ì…, í˜„ì¬ë„ ì¡°ê¸°ìƒí™˜ í‰ê°€ê°€ê²©ì„ ë°‘ëŒê³  ìˆì„ ê°€ëŠ¥ì„±ì´ ìˆë‹¤.
"""
cleaned_content = preprocess_news_content(news_content)
```

```python
from summarizer import Summarizer

def recommend_news(title_query, content_query, top_k=4):
    title_query_embedding = embedder.encode(title_query, convert_to_tensor=True)
    content_query_embedding = embedder.encode(content_query, convert_to_tensor=True)

    query_embedding = torch.cat((title_query_embedding, content_query_embedding), dim=0)

    cos_scores = util.pytorch_cos_sim(query_embedding, corpus_embeddings)[0]
    cos_scores = cos_scores.cpu()

    # ìœ ì‚¬ë„ê°€ 100%ì¸ ê²½ìš°(ì™„ì „íˆ ë™ì¼í•œ ë‰´ìŠ¤)ë¥¼ ì°¾ì•„ ì¸ë±ìŠ¤ë¥¼ ì €ì¥
    same_news_indices = [idx for idx, score in enumerate(cos_scores) if abs(score - 1.0) < 0.00001]

    recommended_news = []
    for idx in range(len(cos_scores)):
        if idx not in same_news_indices:

            recommended_news.append({"title": corpus[idx]["title"].strip(),
                                     "link": corpus[idx]["link"],
                                     "score": cos_scores[idx].item(),
                                     "content": corpus[idx]["content"].strip(),
                                     'content original' : corpus[idx]['content original']
                                     })

    # ìœ ì‚¬ë„(score)ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ì¶”ì²œ ë‰´ìŠ¤ë¥¼ ë‚´ë¦¼ì°¨ìˆœìœ¼ë¡œ ì •ë ¬
    recommended_news = sorted(recommended_news, key=lambda x: x["score"], reverse=True)

    return recommended_news[:top_k]
```
